```python
plt.figure(figsize=(10,6))
df['hotel'].value_counts().plot(kind='bar', color='skyblue')
plt.title('BarChart for hotel')
plt.xlabel('hotel')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['is_canceled'].value_counts().plot(kind='bar', color='red')
plt.title('Barchart for canceled bookings')
plt.xlabel('is_canceled')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['adults'].value_counts().plot(kind='bar', color='red')
plt.title('BarChart for Adults Bookings')
plt.xlabel('adults')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='adults', y='is_canceled', data=df, color='green')
plt.title('Scatterplot for Adults and Canceled booking ')
plt.xlabel('adults')
plt.ylabel('Canceled')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 6))

sns.histplot(df['adults'], bins=40, kde=True, color='skyblue', ax=axes[0])
axes[0].set_title('Number of Adults')
axes[0].set_xlabel('Adults')
axes[0].set_ylabel('Count')

sns.histplot(df['children'], bins=40, kde=True, color='green', ax=axes[1])
axes[1].set_title('Number of Children')
axes[1].set_xlabel('Children')
axes[1].set_ylabel('Count')

sns.histplot(df['babies'], bins=40, kde=True, color='orange', ax=axes[2])
axes[2].set_title('Number of Babies')
axes[2].set_xlabel('Babies')
axes[2].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
