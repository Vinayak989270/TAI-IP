```python
# Scatter plot for Runrate A of Team A
plt.figure(figsize=(10, 6))
plt.scatter(df['Team A'], df['Runrate A'], color='purple', alpha=0.6, edgecolor='black')
plt.title('Scatter Plot of Runrate A for Team A')
plt.xlabel('Team A')
plt.ylabel('Runrate A')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Bar plot of Wining Team
plt.figure(figsize=(12, 6))
Wining_Team = df['Wining Team'].value_counts().head(10)
sns.barplot(x=Wining_Team.index, y=Wining_Team.values, palette='viridis')
plt.title('Wining Team')
plt.xlabel('Wining Team')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['No Balls A'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of No Balls A')
plt.xlabel('No Balls A')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Score A'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Score A')
plt.xlabel('Score A')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
